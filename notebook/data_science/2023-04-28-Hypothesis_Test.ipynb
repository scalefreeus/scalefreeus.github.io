# 비모수적 방법

## 정규성 검정

```python
from scipy import stats
treated	 = [540, 670, 1000,	960, 1200, 4650, 4200]
untreated	= [5000, 4200, 1300, 900, 7400, 4500, 7500]
print(stats.shapiro(treated))
print(stats.shapiro(untreated))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.kdeplot(treated, label='Treated')
sns.kdeplot(untreated, label='Untreated')
plt.legend()
```

## 등분산 검정

```python
r= stats.levene(treated, untreated)
print(r)
```

Levene의 등분산 검정  
$H_0$ : 두 군의 분산은 같다.   
$H_1$ : 두 군의 분산은 같지 않다.   

## Mann-Whitney test

순위합 검정은 일반적으로 독립표본 T 검정에 비해 검정력이 낮으며 순위만 비교한 것이기 때문에 두 군의 크기의 차이 (평균의 차이)를 언급할 수 없는 단점이 있지만 정규분포에 대한 가정을 하지 않기 때문에 크기 순서가 있는 어떤 상황에서도 적용할 수 있는 장점이 있음

귀무가설 $H_0$ : 두 군의 크기가 같다.  
대립가설 $H_1$ : 두 군의 크기가 같지 않다.

```python
u_statistic, p = stats.mannwhitneyu(treated, untreated, alternative="two-sided")
print("Mann-Whitney의 U Score: {}, 유의확률 : {}".format(u_statistic, p))
```

# 모수적 방법

## 정규성 검점

```python
treated	 = [540, 960, 1900,	2500, 2400, 4650, 4200]
untreated	= [5000, 4200, 1300, 900, 7400, 4500, 7500]
print(stats.shapiro(treated))
print(stats.shapiro(untreated))
sns.kdeplot(treated, label='Treated')
sns.kdeplot(untreated, label='Untreated')
plt.legend()
```

## 등분산 검정

```python
r= stats.levene(treated, untreated)
print(r)
```

## T-test
가설검정  
$ H_0 $: 두 군의 평균은 같다  
$ H_1 $: 두 군의 평균은 같지 않다.  

```python
print("두 군의 평균은 같다.", stats.ttest_ind(treated, untreated, equal_var=True))
```
